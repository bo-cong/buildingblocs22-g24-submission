```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("Grades - Sheet1.csv")
```

```python
df.head()
```

```python
df=df[['Topic1','Last Assessment','Days Since Last Studied','Topic2','LA2','DSLS2','Topic3','DSLS3','LA3','TA']]
df=df.rename({"ConvertedVal":"You should study:"},axis=1)
df.head()
df
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
df["Topic1"].value_counts()
```

```python
filt=(df['Days Since Last Studied']=='Never/ALT')
df.loc[filt, 'Days Since Last Studied']=100
filt=(df['DSLS2']=='Never/ALT')
df.loc[filt, 'DSLS2']=100
filt=(df['DSLS3']=='Never/ALT')
df.loc[filt, 'DSLS3']=100
df[['Topic1','Days Since Last Studied','Topic2','LA2','DSLS2','Topic3','DSLS3','LA3','TA']]
```

```python
from sklearn.preprocessing import LabelEncoder
le_t1=LabelEncoder()
df['Topic1']=le_t1.fit_transform(df['Topic1'])

le_t2=LabelEncoder()
df['Topic2']=le_t2.fit_transform(df['Topic2'])

le_t3=LabelEncoder()
df['Topic3']=le_t3.fit_transform(df['Topic3'])

le_ta=LabelEncoder()
df['TA']=le_ta.fit_transform(df['TA'])
df['Days Since Last Studied']=df['Days Since Last Studied'].astype(float)
```

```python
df['Days Since Last Studied']=df['Days Since Last Studied'].astype(float)
df['DSLS2']=df['DSLS2'].astype(float)
df['DSLS3']=df['DSLS3'].astype(float)
```

```python
le_la=LabelEncoder()
df['Last Assessment']=le_la.fit_transform(df['Last Assessment'])
le_la2=LabelEncoder()
df['LA2']=le_la2.fit_transform(df['LA2'])
le_la3=LabelEncoder()
df['LA3']=le_la3.fit_transform(df['LA3'])
```

```python
X=df.drop("TA",axis=1)
y=df['TA']
df['TA']
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(X, y.values)
```

```python
y_pred=logreg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error=np.sqrt(mean_squared_error(y,y_pred))
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dectreereg=DecisionTreeRegressor(random_state=0)
dectreereg.fit(X,y.values)
y1_pred=dectreereg.predict(X)
error=np.sqrt(mean_squared_error(y,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
max_depth = [None, 2,4,6,8,10,12]
parameters = {'max_depth':max_depth}
regressor=DecisionTreeRegressor(random_state=0)
gs=GridSearchCV(regressor,parameters,scoring='neg_mean_squared_error')
gs.fit(X,y.values)
```

```python
regressor=gs.best_estimator_
regressor.fit(X, y.values)
y_pred=regressor.predict(X)
X
```
